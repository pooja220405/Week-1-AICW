```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('house_dataset.csv')
data.head()
```

```python
sns.scatterplot(x=data['Area'],y=data['Price'])
plt.show()
```

```python
sns.scatterplot(x=data['Age'], y=data['Price'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
#separate independent variables
X = data.drop('Price',axis='columns')
X.head()
```

```python
#separate dependent variables
Y = data['Price']
Y.head()
```

```python
#train
reg.fit(X,Y)
```

```python
reg.predict([[2000,3,25]])
```

```python
reg.intercept_#c value only one
```
