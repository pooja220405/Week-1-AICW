```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('homeprices.csv')
data.head()
```

```python
sns.scatterplot(x='area', y='price',data=data,color='purple',marker='*')
plt.show()
```

```python
#import linear regression model
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
#separate independent and dependent features
x= data.drop('price',axis='columns')#independent feature
x.head()
```

```python
y=data.price
y.head()
```

```python
#train linear regression model
reg.fit(x,y)
```

```python
reg.predict([[2000]])
```

```python
reg.intercept_# c value
```

```python
reg.coef_ #m value
```

```python
Price = reg.coef_ * 2000 + reg.intercept_
Price
```

```python
Area = int(input("Enter Area :"))
Price = reg.coef_* Area + reg.intercept_
print("Price of the house is :",Price)
```
