```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#read dataset
data = pd.read_csv('insurance_data.csv')
data.head()
```

```python
sns.scatterplot(x='age',y='bought_insurance',data=data)
plt.show()
```

```python
x = data.drop("bought_insurance",axis=1)
x.head()
```

```python
y = data.bought_insurance
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression()
```

```python
LR_model.fit(x_train,y_train)
```

```python
LR_model.predict([[50]])
```

```python
LR_model.score(x_train,y_train)
```

```python
LR_model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
 
```

```python
y_train_predict = LR_model.predict(x_train)
y_train_predict
```

```python
cm_train_data = confusion_matrix(y_train,y_train_predict)
cm_train_data
```

```python
y_test_predict = LR_model.predict(x_test)
y_test_predict
```

```python
cm_test_data = confusion_matrix(y_test,y_test_predict)
cm_test_data
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
disp_train = ConfusionMatrixDisplay(cm_train_data,display_labels=[0,1])
disp_train.plot()
plt.show()
```

```python
disp_test = ConfusionMatrixDisplay(cm_test_data,display_labels=[0,1])
disp_test.plot()
plt.show()
```
