```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
plt.plot(df.area, df.price)
plt.show()
```

```python
plt.scatter(df.area, df.price)
plt.show()
```

```python
X = df.drop("price", axis="columns") #independent feature
X.head()
```

```python
y = df.price #dependent feature
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
lin_reg.predict([[3300]])
```

```python
lin_reg.predict([[1230]])
```

```python
lin_reg.coef_ #slope m
```

```python
lin_reg.intercept_ #intercept b
```

```python
price = lin_reg.coef_ *3300 + lin_reg.intercept_
price
```
